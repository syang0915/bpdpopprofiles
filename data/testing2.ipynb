```python
import pandas as pd
import os
```

```python
employees_2025  = pd.read_csv("bpi_employee_roster_fa_25.csv")
employees_2025.head()
employees_2025_short = employees_2025[["employee_id", "first_name", "last_name"]]

def extract_middle_name(first_name):
    parts = str(first_name).split()
    if len(parts) > 1:
        return ' '.join(parts[1:])
    return ''

def extract_first_name(first_name):
    parts = str(first_name).split()
    return parts[0] if parts else ''

employees_2025_short['middle_name'] = employees_2025_short['first_name'].apply(extract_middle_name)
employees_2025_short['first_name'] = employees_2025_short['first_name'].apply(extract_first_name)
employees_2025_short.head()
```

```python
employees_all = pd.read_csv("incident_with_employee_ids.csv")
employees_all.head()
```

```python
bpi_police_financial = pd.read_csv("bpi_police_financial.csv")
employees_2025_short_joined = employees_2025_short.merge(bpi_police_financial, on="employee_id", how="left")
employees_2025_short_joined.head(10)
```

```python
# Join with employees_all and keep only specified columns
final_df = employees_2025_short_joined.merge(employees_all, left_on="employee_id", right_on="Employee ID", how="left")

# Select only the desired columns
salary_cols = ['regular_pay', 'retro_pay', 'other_pay', 'ot_pay', 'injured_pay', 'detail_pay', 'quinn_pay', 'total_pay']
cols_to_keep = ['first_name_x', 'last_name_x', 'middle_name', 'employee_id', 'Zip Code'] + salary_cols

final_df_selected = final_df[cols_to_keep].copy()
final_df_selected.columns = ['first_name', 'last_name', 'middle_name', 'employee_id', 'zip_code', 'regular_pay', 'retro_pay', 'other_pay', 'ot_pay', 'injured_pay', 'detail_pay', 'quinn_pay', 'total_pay']

final_df_selected
```

```python
# Split into two separate tables
# Table 1: Employee information
employees_table = final_df_selected[['employee_id', 'first_name', 'last_name', 'middle_name', 'zip_code']].drop_duplicates()
print("Employee Information Table:")
employees_table.shape
```

```python
# Table 2: Salary information
salary_table = final_df_selected[['employee_id', 'regular_pay', 'retro_pay', 'other_pay', 'ot_pay', 'injured_pay', 'detail_pay', 'quinn_pay', 'total_pay']]
print("\nSalary Information Table:")
salary_table
```

```python
# Save to separate CSV files without index
employees_table.to_csv('employees_2025_info.csv', index=False)
salary_table.to_csv('employees_2025_salary.csv', index=False)

print("Saved employees_2025_info.csv")
print("Saved employees_2025_salary.csv")
```

```python
# Create all-years tables using bpi_police_financial data
# First, merge police financial with employees_all to get complete info
all_years_data = bpi_police_financial.merge(employees_all[['Employee ID', 'First_Name', 'Middle_Initial']], left_on='employee_id', right_on='Employee ID', how='left')

# Table 1: All employee information (unique employees)
all_employees_table = all_years_data[['employee_id', 'first_name', 'last_name', 'Zip Code']].drop_duplicates()
all_employees_table.columns = ['employee_id', 'first_name', 'last_name', 'zip_code']

# Table 2: All salary information (with year)
all_salary_table = all_years_data[['employee_id', 'year', 'regular_pay', 'retro_pay', 'other_pay', 'ot_pay', 'injured_pay', 'detail_pay', 'quinn_pay', 'total_pay']]

# Save to CSV files without index
all_employees_table.to_csv('employees_all_info.csv', index=False)
all_salary_table.to_csv('employees_all_salary.csv', index=False)

print("Saved employees_all_info.csv")
print("Saved employees_all_salary.csv")
print(f"Total employees: {len(all_employees_table)}")
print(f"Total salary records: {len(all_salary_table)}")
```
