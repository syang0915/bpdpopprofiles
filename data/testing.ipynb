```python
def split_cell(cell):
    if pd.isna(cell):
        return []
    return [x.strip() for x in str(cell).splitlines() if x.strip()]

new_rows = []

for _, row in df.iterrows():
    split_data = {col: split_cell(row[col]) for col in multi_cols}

    n = max((len(v) for v in split_data.values()), default=0)
    if n == 0:
        n = 1

    for i in range(n):
        new_row = {col: row[col] for col in shared_cols}
        for col in multi_cols:
            values = split_data[col]
            new_row[col] = values[i] if i < len(values) else None
        new_rows.append(new_row)

clean_df = pd.DataFrame(new_rows)
clean_df.to_csv("normalized_cases.csv", index=False)
```
